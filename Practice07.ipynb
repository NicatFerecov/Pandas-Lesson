```python
def longest_durations(df, top_n=5):
    movies = df[df['type'] == 'Movie'].copy()
    movies['duration_min'] = movies['duration'].str.replace(' min', '', regex=False).astype(float)
    top_movies = movies.sort_values(by='duration_min', ascending=False).head(top_n)

    shows = df[df['type'] == 'TV Show'].copy()
    shows['seasons'] = shows['duration'].str.replace(' Season', '', regex=False).str.replace('s', '', regex=False).astype(float)
    top_shows = shows.sort_values(by='seasons', ascending=False).head(top_n)

    return top_movies[['title', 'duration_min']], top_shows[['title', 'seasons']]


longest_durations(df)
```

```python
def longest_titles(df, top_n=5):
    df['title_length'] = df['title'].str.len()
    longest = df.sort_values(by='title_length', ascending=False).head(top_n)
    shortest = df.sort_values(by='title_length').head(top_n)
    return longest[['title', 'title_length']], shortest[['title', 'title_length']]
longest_titles(df)
```

# Conclusion:

```python
import pandas as pd
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
import re

class NetflixAdvancedAnalyzer:
    def __init__(self, df):
        self.df = df.copy()
        self.df['date_added'] = pd.to_datetime(self.df['date_added'].astype(str).str.strip(), errors='coerce')
        self.df['release_year'] = pd.to_numeric(self.df['release_year'], errors='coerce')
        self.df['year_added'] = self.df['date_added'].dt.year

    def release_vs_added_gap(self):
        self.df['delay'] = self.df['year_added'] - self.df['release_year']
        delays = self.df['delay'].dropna()
        return delays.describe(), delays.value_counts().sort_index()

    def multi_category_content(self):
        df_clean = self.df.dropna(subset=['listed_in'])
        df_clean['category_count'] = df_clean['listed_in'].apply(lambda x: len(x.split(',')))
        multi = df_clean[df_clean['category_count'] > 1]
        return multi[['title', 'listed_in', 'category_count']].sort_values(by='category_count', ascending=False)

    def popular_category_combinations(self, top_n=10):
        all_combos = self.df['listed_in'].dropna().apply(lambda x: tuple(sorted(x.split(','))))
        combo_counts = Counter(all_combos)
        return combo_counts.most_common(top_n)

    def common_words_in_titles(self, top_n=10):
        words = self.df['title'].dropna().apply(lambda x: re.findall(r'\w+', x))
        all_words = [word.lower() for sublist in words for word in sublist]
        word_counts = Counter(all_words)
        return word_counts.most_common(top_n)

    def delayed_content(self, threshold=5):
        self.df['delay'] = self.df['year_added'] - self.df['release_year']
        anomalies = self.df[self.df['delay'] >= threshold]
        return anomalies[['title', 'release_year', 'year_added', 'delay']].sort_values(by='delay', ascending=False)

    def plot_release_year_distribution(self):
        counts = self.df['release_year'].value_counts().sort_index()
        plt.figure(figsize=(12,6))
        counts.plot(kind='bar')
        plt.title('Yapım Yılına Göre İçerik Sayısı')
        plt.xlabel('Yapım Yılı')
        plt.ylabel('İçerik Sayısı')
        plt.grid(axis='y')
        plt.show()

    def longest_durations(self, top_n=5):
        movies = self.df[self.df['type'] == 'Movie'].copy()
        movies['duration_min'] = movies['duration'].str.replace(' min', '', regex=False).astype(float)
        top_movies = movies.sort_values(by='duration_min', ascending=False).head(top_n)

        shows = self.df[self.df['type'] == 'TV Show'].copy()
        shows['seasons'] = shows['duration'].str.replace(' Season', '', regex=False).str.replace('s', '', regex=False).astype(float)
        top_shows = shows.sort_values(by='seasons', ascending=False).head(top_n)

        return top_movies[['title', 'duration_min']], top_shows[['title', 'seasons']]

    def most_added_days(self, top_n=5):
        added_counts = self.df['date_added'].value_counts().head(top_n)
        return added_counts

    def single_country_content(self):
        df_clean = self.df.dropna(subset=['country'])
        single_country = df_clean[~df_clean['country'].str.contains(',', na=False)]
        return single_country[['title', 'country']]
```
