````python
import pandas as pd

data = {
    'product': ['Laptop', 'Phone', 'Tablet', 'Monitor', 'Keyboard', 'Mouse', 'Printer'],
    'price': [1200, 800, 300, 400, 100, 50, 500],
    'category': ['Electronics', 'Electronics', 'Electronics', 'Electronics', 'Accessories', 'Accessories', 'Electronics'],
    'discount': [10, 5, 20, 15, 0, 0, 25],
    'in_stock': [True, True, False, True, True, False, True]
}

df = pd.DataFrame(data)
````

````python
df[df['price'] > 500]
df[df['category'] == 'Accessories']
df[(df['price']  > 500) & (df['discount'] >= 10)]
df[df['in_stock']]
df[df['discount'] == 0]
df[(df['price'] >= 300) & (df['price'] <= 1000)]
df[(df['category'] == "Electronics") & (df['in_stock'])]
df['price'].max()
df[df['discount'].isin([10, 15, 20])]
````
