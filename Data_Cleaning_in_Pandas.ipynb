# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel('/content/Customer Call List.xlsx')
```

```python
df
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df.drop(columns='Not_Useful_Column', inplace=True)
```

```python
df
```

```python
df['Last_Name'] = df['Last_Name'].str.replace("/", "")
```

```python
df['Last_Name'] = df['Last_Name'].str.strip("_")
df['Last_Name'] = df['Last_Name'].str.strip("_")
```

```python
df
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('...')
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','')
```

```python
df
```

```python
df.dtypes
```

```python
# str(df['Phone_Number'])
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('N-a--', '')
```

```python
df
```

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(pat=',', n=2, expand=True)
```

```python
df.drop(columns='Address', inplace=True)
```

```python
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
```

```python
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
```

```python
df
```

```python
df['Paying Customer'].str.strip("/a")
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
```

```python
df
```

```python
df = df.replace('N/a', '', regex=True)
```

```python
df
```

```python
df.fillna('', inplace=True)
```

```python
df
```

```python
for x in df.index:
  if df.loc[x, 'Do_Not_Contact'] == 'Y':
    df.drop(x, inplace=True)

df
```

```python
for x in df.index:
  if df.loc[x, 'Phone_Number'] == '':
    df.drop(x, inplace=True)

df

df.dropna(subset='Phone_Number', inplace=True)
```

```python
df = df.reset_index(drop=True)
df
```
