# Group By and Aggregate Functions in Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Flavors.csv')
```

```python
df
```

```python
group_by_frame = df.groupby('Base Flavor')
```

```python
group_by_frame.mean(numeric_only=True)
```

```python
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').min()
```

```python
df.groupby('Base Flavor').max()
```

```python
df.groupby('Base Flavor').sum()
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating': ['max', 'count', 'mean', 'sum'], 'Texture Rating': ['mean', 'max', 'count', 'sum']})
```

```python
df.groupby(['Base Flavor','Liked']).sum()
```

```python
df.groupby(['Base Flavor','Liked']).agg({'Flavor Rating': ['max', 'count', 'mean', 'sum']})
```

```python
df.groupby('Base Flavor').describe()
```
