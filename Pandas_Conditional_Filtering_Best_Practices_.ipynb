```python
import pandas as pd

data = {
    'Product': ['Phone Pro', 'Laptop Air', 'Phone Lite', 'Smart TV', 'Tablet Pro'],
    'Price': [950, 1200, 600, 800, 700],
    'Category': ['Electronics', 'Electronics', 'Electronics', 'Electronics', 'Electronics'],
    'Brand': ['Apple', 'Apple', 'Xiaomi', 'Samsung', 'Apple']
}

df = pd.DataFrame(data)
```

```python
df[df['Price'] > 800]
```

```python
df[(df['Brand'] == "Apple") & (df['Category'] == "Electronics") & (df['Price'] > 700)]
```

```python
df[df['Brand'].isin(['Samsung'])]
```

```python
df[df['Price'].between(100, 700)]
```

```python
df[df['Product'].str.contains("TV")]
```
