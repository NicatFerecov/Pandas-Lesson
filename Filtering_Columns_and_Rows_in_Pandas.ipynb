# Filtering Columns and Rows in Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/StudentsPerformance.csv')
```

```python
df
```

```python
specific_education = ["master's degree", "high school"]
```

```python
df[df['parental level of education'].isin(specific_education)]
```

```python
df[df['test preparation course'].str.contains('completed')]
```

```python
df2 = df.set_index('gender')
df2
```

```python
df2.filter(items=['math score', 'reading score'], axis=1)
```

```python
df2.filter(items = ['group B'], axis=1)
```

```python
df.filter(like = 'completed', axis=1)
```

```python
df2.loc["female"]
```

```python
df2.iloc[3]
```

# Order by

```python
df[df['writing score'] > 30].sort_values(by=['gender', 'parental level of education'], ascending=[False, True])
```
